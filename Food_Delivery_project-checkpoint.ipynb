# Food Delivery Dataset


About Dataset


Food delivery is a courier service in which a restaurant, store, or independent food-delivery company delivers food to a customer. An order is typically made either through a restaurant or grocer's website or mobile app, or through a food ordering company. The delivered items can include entrees, sides, drinks, desserts, or grocery items and are typically delivered in boxes or bags. The delivery person will normally drive a car, but in bigger cities where homes and restaurants are closer together, they may use bikes or motorized scooters

```python
# Renaming the colunm name Time_taken(min) to Time_taken
df.rename(columns = {'Time_taken(min)':'Time_taken'}, inplace = True)
```

```python
(df['Time_Orderd'].unique())
```

```python
df_cleaned = df.dropna(subset=['Time_Orderd'])
```

```python
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str)
        return True
    except ValueError:
        return False
```

```python
df_cleaned = df_cleaned[df_cleaned['Time_Orderd'].apply(is_valid_date)]
```

```python
df_cleaned['Time_Orderd'] = pd.to_datetime(df_cleaned['Time_Orderd'])
```

```python
#Converting (Order_Date) column type (Object) to (Date and time) because we have to perfrom anlyasis with this column.

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
# Distribution plot with Delivery_person_Ratings column
plt.figure(figsize=(10,10))
sns.countplot(data=df, x="Delivery_person_Ratings")
plt.show()
```

```python
# Scatter plot with rating based on Age
plt.figure(figsize=(15, 6))
sns.scatterplot(x='Delivery_person_Age', y='Delivery_person_Ratings', data=df)
plt.title('Scatter Plot of Ratings and Age')
plt.xlabel('Age')
plt.ylabel('Ratings')
plt.show()
```

```python
# Plot histogram for 'age'
plt.figure(figsize=(10, 6))
sns.histplot(df['Delivery_person_Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Delivery Times by Weather Conditions
plt.figure(figsize=(10, 6))
sns.boxplot(x='Weatherconditions', y='Time_taken', data=df_cleaned)
plt.title('Delivery Times by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Time Taken')
plt.show()
```

```python
# Delivery Times by Traffic Density
plt.figure(figsize=(10, 6))
sns.boxplot(x='Road_traffic_density', y='Time_taken', data=df_cleaned)
plt.title('Delivery Times by Traffic Density')
plt.xlabel('Road Traffic Density')
plt.ylabel('Time Taken')
plt.show()
```

```python
# Order Volume by Time of Day
df_cleaned['Hour_of_Day'] = df_cleaned['Time_Orderd'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour_of_Day', data=df_cleaned)
plt.title('Order Volume by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Order Count')
plt.show()
```
